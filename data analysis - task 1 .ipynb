反斜杠是转义符，意思是它被用来表示特殊字符，比如换行符 \n

如果字符串中包含许多反斜杠但没有特殊字符，可以在字符串前面加一个r，表明字符就是他自身

```python
s = r'this\\has\no\characters'
s
```
